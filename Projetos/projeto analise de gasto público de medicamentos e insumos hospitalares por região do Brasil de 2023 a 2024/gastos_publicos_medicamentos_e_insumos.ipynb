# Outliers

```python
plt.figure(figsize=(10, 6))                                                     # Visualizando a distribuição dos dados com outliers
plt.boxplot(df_combinado["Preço Total"], vert=False, whis=1.5)
plt.title("Boxplot de preços totais")
plt.xlabel("Preço Total")
plt.show()
```

```python

Q1 = df_combinado["Preço Total"].quantile(0.25)         # Identificando os Outliers
Q3 = df_combinado["Preço Total"].quantile(0.75)
IQR = Q3 - Q1  # Intervalo Interquartil

# Limites para detectar outliers
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

print("Q1:", Q1)
print("Q3:", Q3)
print("IQR:", IQR)
print(f"Limite Inferior (outliers): {limite_inferior}")
print(f"Limite Superior (outliers): {limite_superior}")
```

```python
df_sem_outliers = df_combinado[
    (df_combinado["Preço Total"] >= limite_inferior) &
    (df_combinado["Preço Total"] <= limite_superior)
]
```

```python
quartis_sem_outliers = df_sem_outliers["Preço Total"].quantile([0.25, 0.5, 0.75])
quartis_sem_outliers
```

```python
plt.figure(figsize=(10, 6))                                                     # Visualizando a distribuição dos dados sem outliers
plt.boxplot(df_sem_outliers["Preço Total"], vert=False, whis=1.5)
plt.title("Boxplot de preços totais (sem outliers)")
plt.xlabel("Preço Total")
plt.show()
```
